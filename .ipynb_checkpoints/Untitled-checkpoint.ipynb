```python
# importing required modules
from PyPDF2 import PdfReader
  
# creating a pdf reader object
reader = PdfReader('01_22.pdf')
  
# printing number of pages in pdf file
print(len(reader.pages))
  
# getting a specific page from the pdf file
page = reader.pages[0]
  
# extracting text from page
pdf_text = page.extract_text()
print(pdf_text)
```

```python
import re
import pandas as pd
 
# Separa o texto em linhas
lines = pdf_text.strip().split('\n')

# Lista para armazenar os dados da tabela
table_data = []

# Encontra a linha que contém os cabeçalhos da tabela
header_row = None
for i, line in enumerate(lines):
    if "Produtos" in line and "Medidas" in line and "Valor" in line:
        header_row = i
        break

# Encontra as linhas de dados da tabela
if header_row is not None:
    data_rows = lines[header_row + 1:]
    for row in data_rows:
        # Usa regex para dividir a linha em colunas
        columns = re.split(r'\s{2,}', row.strip())
        # Remove os valores vazios da lista de colunas
        columns = [col for col in columns if col]
        table_data.append(columns)

# Cria o DataFrame a partir dos dados da tabela
df = pd.DataFrame(table_data, columns=["Produtos", "Medidas", "Valor Mínimo", "Valor Máximo", "Valor Médio", "Valor Comum", "Valor Mediana"])

print(df)
```
